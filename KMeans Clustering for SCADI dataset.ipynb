### Q2 For the dataset, perform PCA. 
a) Plot the captured variance with respect to increasing latent dimensionality. 

b) What is the minimum dimension that captures : 
 -- At least 89% variance 
 -- At least 99% variance 

```python
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn import metrics as skmet

#normalize our data
Xnorm = scale(scadi1)

#perform PCA using sklearn PCA implementation
pca = PCA(n_components=70)
pca.fit(Xnorm)
print(pca)

print('*'*50)

#The amount of variance that each PC explains
var= pca.explained_variance_ratio_
print(np.round(var*100,2))

print('*'*50)

#Cumulative Variance explains
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
print(var1)
plt.plot(var1)
plt.xlabel("Principal components")
plt.ylabel("Variance captured")
plt.show()

print('*'*50)

pca = PCA(n_components=70)
Zred = pca.fit_transform(Xnorm)
print(Zred.shape)

print('*'*50)

# Reconstruct our data
Xrec = pca.inverse_transform(Zred)
print(Xrec.shape)

print('*'*50)

#Visualize the recovered data
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(18,5))  # added size of each figs (width, height)
fig.subplots_adjust(wspace=0.2) # leave some space between figs


# plot for Xnorm 
axs[0].scatter(Xnorm[:,0], Xnorm[:,1])
axs[0].set_title("Normalised Original Data")


# plot for Xrec
axs[1].scatter(Xrec[:,0], Xrec[:,1])
axs[1].set_title("Recovered data after dimensionality reduction")
plt.show()

print('*'*50)

# Measure the reconstruction error
rec_error = np.linalg.norm(Xnorm-Xrec, 'fro')/np.linalg.norm(Xnorm, 'fro')
print(f'The reconstruction error is: {rec_error}')
```

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# output charts 
%matplotlib inline

# Preprocessing
from sklearn.preprocessing import StandardScaler

# Models for classification 
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
# Model evaluation
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
disease
```

```python
#Splitting data into x (target) and y (predictor variables)
x = disease.drop('num', axis=1)
y = disease['num']
x
```

```python
y
```

```python
#Training and test dataset
np.random.seed(100)

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
#Use StandardScaler for SVG and optimise the training
scaler1 = StandardScaler().fit(X_train)
scaler1.scale_
```

```python
#Reference : https://github.com/JorgePasco1/heart-disease-classification/blob/master/heart-disease-classification.ipynb

Model = {'Random Forest' : RandomForestClassifier(),
        'Logistic Regression' : LogisticRegression(),
        'KNeighbours' : KNeighborsClassifier(),
        'LinearSVC' : LinearSVC()}

# We'll create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):

    #setting the random seed 
    np.random.seed(100)

    #store the scores
    Metrics = {}

    #Make a loop through the models - 
    for name, model in Model.items():
        #Adjusting for Linear SVC
        X_train_ad = scaler1.transform(X_train) if name == 'LinearSVC' else X_train
        X_test_ad = scaler1.transform(X_test) if name == 'LinearSVC' else X_test
    
        #Model fit 
        model.fit(X_train_ad, y_train)
    
        Metrics[name] = model.score(X_test_ad, y_test)
    
    return Metrics
```

```python
Metrics = fit_and_score(models = Model,
                        X_train=X_train,
                        X_test=X_test,
                        y_train=y_train,
                        y_test=y_test)
Metrics
```

```python
Diff_models = pd.DataFrame(Metrics, index=['accuracy'])
Diff_models.T.plot(kind='bar', color='green');
```

```python
reg = Metrics['Logistic Regression']['Model']
y_pred = reg.predict(X_test)
y_pred
```
