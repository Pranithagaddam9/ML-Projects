```python
# Business case : On the basis of given parameters (features) our
# target is to find 
## whether a patient will have B type of M type cancer. 
# We are solving a classification task. 
```

```python
# importing the libraries 
import warnings
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
import pandas as pd
```

```python
data = pd.read_csv('breast cancer.csv')
```

```python
print(data.shape)
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(['id'],axis=1,inplace=True)
```

```python
data.head(5)
```

```python
## creating independent and dependent variables
X = data.iloc[:,1:]
y = data.diagnosis
```

### Transform X and y

```python
## Scaling the data as the magnitude of variables are varying

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
scaled_x = sc.fit_transform(X)
X = pd.DataFrame(scaled_x, columns=X.columns)
```

```python
from sklearn.preprocessing import LabelEncoder

label_y = LabelEncoder()

label_y.fit_transform(y)
```

```python
## creating training and test data 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=0.2,random_state=10)
```

```python
# model creation - Fully connected layers

from sklearn.neural_network import MLPClassifier
model = MLPClassifier( hidden_layer_sizes = (50,3), #3 sets of 50 neurons
                      learning_rate_init=0.1, #gradient descent algorithm
                      max_iter=100,
                      random_state=2) ## model object creation

model.fit(X_train, y_train) ##training the data 
y_predict_proba = model.predict_proba(X_test) #predicting probability of class
y_predict = model.predict(X_test)
y_train_predict = model.predict(X_train)

#Forward propagation - Summation and Activation
#Backward propagation - updating the optimal values using gradient descent
# 1st iteration - [FP + BP + errors] = epochs
# 2nd iteration - [FP + BP + errors] = epochs 
# 100 iterations for each hidden layers 
```

```python
# Evaluating the model created 

from sklearn.metrics import accuracy_score, confusion_matrix

print("Train accuracy :", accuracy_score(y_train, y_train_predict))

print("Test accuracy :", accuracy_score(y_test, y_predict)) #test accuracy
```

```python
# Confusion matrix
pd.crosstab(y_test,y_predict)
```
